```python
import io
import requests
import json

url = "https://scop.mrc-lmb.cam.ac.uk/api/ancestry/4000345"
js = requests.get(url).content
parsed = json.loads(js)
print(json.dumps(parsed, indent=0, sort_keys=True))
```

```python
import pandas as pd
import io
import requests

url = "https://api.covidtracking.com/v1/us/daily.csv"

data = requests.get(url).content
csv = pd.read_csv(io.StringIO(data.decode("utf-8")))
print(csv)
```
